```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv('/content/drive/MyDrive/seasons_1234(updated).csv')
```

```python
df.head(5)
```

Preprocessing:

```python
# firstly, let`s decapitalize all the names and lines:
df.name = df['name'].str.lower()
df.line = df['line'].str.lower()
df.name = df['name'].str.split(' ').str[0]
```

```python
df.name.unique()
```

As we can see here, the column "name" is a bit messy, names and surnames are written separately (e.g. Jake and Peralta is one person)

```python
df['name'] = df.name.replace('boyle','charles')
df['name'] = df.name.replace('peralta','jake')
df['name'] = df.name.replace('captain','holt')
df['name'] = df.name.replace('santiago','amy')
df['name'] = df.name.replace('diaz','rosa')
```

```python
# Continue with loading all necessary libraries
from os import path
from PIL import Image
from wordcloud import WordCloud, ImageColorGenerator

import matplotlib.pyplot as plt
```

```python
# For stopwords removing I chose Spacy:
import spacy
# loading the english language small model of spacy
en = spacy.load('en_core_web_sm')
stops = en.Defaults.stop_words
```

Now let`s create a wordcloud for Jake:

```python
jake = " ".join(line for line in df[df["name"]=="jake"].line)
wordcloud_jake = WordCloud(stopwords=stops, max_words=100, background_color="white").generate(jake)
# Display the generated image:
# the matplotlib way:
plt.imshow(wordcloud_jake, interpolation='bilinear')
plt.axis("off")
plt.show()
```

```python
amy = " ".join(line for line in df[df["name"]=="amy"].line)
wordcloud_amy = WordCloud(stopwords=stops, max_words=100, background_color="white").generate(amy)
plt.imshow(wordcloud_amy, interpolation='bilinear')
plt.axis("off")
plt.show()
```
